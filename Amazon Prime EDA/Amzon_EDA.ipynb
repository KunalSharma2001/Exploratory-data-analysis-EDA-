As we can see in the above, Table/Data. There are many **Nan/Null** values in the data. Thus we need to replace these NULL values using Various Methods.

-> Here we can see for the following:

    1. Director Column : we cannot put any of the mathematical value by any means i.e taking mean or anything, thus we simply put **Unavailable** inplace of NaN.
    
    2. Cast Column : Similar to the Director column, we will put unavailable inplace of NaN.
    
    
    3. Rating Column : We will the Raiting Column bt the mode of the values of the column.

```python
data['director'] = data['director'].replace(np.nan, "Unavailable")
```

```python
data['cast'] = data['cast'].replace(np.nan, "Unavailable")
```

```python
data['rating'] = data['rating'].fillna(data['rating'].mode()[0])
```

```python
data.head(10)
```

```python
data['country'].isnull().value_counts()
```

Here we can see that, around **8996** values in **Country** Column are NULL.
Thus we need to replace them or Fill them.

We can see that, most of the countries in the data are, Canada, UK, US, India. We can choose
either of these to fill the missing values.

```python
import random
countries = ['India', 'Canada', 'UnitedStates', 'UnitedKingdom']
data['country'] = data['country'].replace(np.nan, random.choice(countries))
```

```python
data['country'].isnull().value_counts()
```

Here we see that all the Missing Values in the Column **Country** are filled.

We can see that in the Column, **Data**,  there are NULL values too.

```python
data['date_added'].isnull().value_counts()
```

Here, we see that tere are 9513, values that are NULL in the Columns = data_added.
Thus we fill these values using **Forward Fill {ffill}**.

```python
data['date_added'] = data['date_added'].ffill()
```

```python
data['date_added'].isnull().value_counts()
```

Thus, now We see there are no Missing Values in the Column.
