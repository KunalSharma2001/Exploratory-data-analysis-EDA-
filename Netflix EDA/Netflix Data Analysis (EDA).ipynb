```python
# Finding out the different type of Genres
# Dividing the dataset into, two set for movies and series.
# file_movie, file_series
file_movie = file[file['Series or Movie'] == 'Movie']
file_movie.head(2)
```

```python
file_series = file[file['Series or Movie'] == 'Series']
file_series.head(2)
```

We filter a movies or a series on the basis of Genre, thus here we will calculate the different types of `Genres` in the given dataset.

```python
file_series['Genre'].unique()
```

```python
# dropping the rows which have NULL values in the `Genre` section
file_series_d = file_series.dropna(subset=['Genre'])
```

```python
series_genre_list = file_series_d.Genre.str.split(',')
s_gen_list = {} # creting an empty list to store the Genre

for series in series_genre_list:
    for genre in series:
        if(genre in series_genre_list):
            s_gen_list[genre] += 1
        else:
            s_gen_list[genre] = 1

series_gen_file = pd.DataFrame(s_gen_list.values(), index = s_gen_list.keys(),
                               columns={'Counts of Genre in TV Series'})
series_gen_file.sort_values(by = 'Counts of Genre in TV Series', ascending=False, inplace=True)
```

```python
file_series_gen = file.dropna(subset=['Genre'])
```

```python
file_list = file_series_gen.Genre.str.split(',')
file_genList = {}

for genres in file_list:
    for genre in genres:
        if(genre in file_genList):
            file_genList[genre]+=1
        else:
            file_genList[genre]=1
            
file_genList = pd.DataFrame(file_genList.values(), index=file_genList.keys(), columns = {'Counts of Genres in TV series'})
```

```python
file_genList.sort_values(by = 'Counts of Genres in TV series',ascending=False, inplace=True)
```

```python
top_10_file_genList = file_genList[:10]
top_10_file_genList
```

```python
# Plotting the figures
plt.figure(figsize=(15,7))
plt.title("Top Genres in Series Category")

sns.barplot(x=top_10_file_genList.index, y=top_10_file_genList['Counts of Genres in TV series'],)
```
